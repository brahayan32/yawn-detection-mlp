# 03 - Entrenamiento MLP

Entrena el perceptron multicapa con TensorFlow puro, sin Keras.

```python
from pathlib import Path
import sys
import numpy as np
import tensorflow as tf

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.config import BATCH_SIZE, DATASET_DIR, METRICS_DIR, MODELS_DIR, EPOCHS, RANDOM_STATE, L2_REGULARIZATION
from src.preprocessing import load_all_dataset
from src.model import PureTensorFlowMLP, save_model, train_mlp
from src.evaluation import save_history_plots
```

```python
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)
data = load_all_dataset(DATASET_DIR)
x_train, y_train, _ = data['train']
x_val, y_val, _ = data['validation']
x_train.shape, y_train.shape, x_val.shape, y_val.shape
```

```python
model = PureTensorFlowMLP(input_dim=x_train.shape[1], seed=RANDOM_STATE)
print('Entrada:', x_train.shape[1])
print('Arquitectura: 6400 -> 256 ReLU -> 64 ReLU -> 1 Sigmoid')
```

```python
history = train_mlp(
    model,
    x_train,
    y_train,
    x_val=x_val,
    y_val=y_val,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    l2_strength=L2_REGULARIZATION,
)

save_history_plots(history, METRICS_DIR)
save_model(model, MODELS_DIR / 'best_model')
```
