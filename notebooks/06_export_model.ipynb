# 06 - Exportacion del modelo

Guarda el modelo final para que el equipo de backend pueda integrarlo.

```python
from pathlib import Path
import sys
import tensorflow as tf

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.config import MODELS_DIR
```

```python
best_model_path = MODELS_DIR / 'best_model'
final_model_path = MODELS_DIR / 'final_model'

model = tf.saved_model.load(str(best_model_path))
tf.saved_model.save(model, str(final_model_path))
final_model_path
```

Para backend, se debe conservar el mismo preprocesamiento: grises, Gaussiano, 64x64, normalizacion y vectorizacion.
